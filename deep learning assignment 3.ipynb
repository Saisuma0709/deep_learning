```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

# Load dataset
data = pd.read_csv('AIQ.csv')  # Replace with your dataset path
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

# Using all features ('Open', 'High', 'Low', 'Close', 'Volume')
features = data[['Open', 'High', 'Low', 'Close', 'Volume']].values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_features = scaler.fit_transform(features)

# Creating sequences for RNN input
def create_sequences(data, look_back=30, target_col=3):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back])
        y.append(data[i + look_back, target_col])  # Predict 'Close' column
    return np.array(X), np.array(y)

look_back = 30  # Using the past 30 days
X, y = create_sequences(scaled_features, look_back)

# Spliting the dataset into training (80%) and testing (20%) sets
train_size = int(0.8 * len(X))
X_train_full, y_train_full = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Further spliting the dataset into training into training and validation (80-20 split)
val_size = int(0.2 * len(X_train_full))
X_train, y_train = X_train_full[:-val_size], y_train_full[:-val_size]
X_val, y_val = X_train_full[-val_size:], y_train_full[-val_size:]

# Building and training models
def build_and_train_model(model_type, input_shape, X_train, y_train, X_val, y_val, learning_rate, dropout_rate, epochs, batch_size):
    model = Sequential()
    if model_type == 'RNN':
        model.add(SimpleRNN(50, activation='tanh', return_sequences=False, input_shape=input_shape))
    elif model_type == 'LSTM':
        model.add(LSTM(50, return_sequences=False, input_shape=input_shape))
    elif model_type == 'GRU':
        model.add(GRU(50, return_sequences=False, input_shape=input_shape))
    
    model.add(Dropout(dropout_rate))
    model.add(Dense(25))
    model.add(Dense(1))
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, batch_size=batch_size, verbose=1)
    return model, history

# Hyperparameter tuning grid
hyperparams = {
    'learning_rate': [0.001, 0.01],
    'dropout_rate': [0.2, 0.3],
    'batch_size': [16, 32],
    'epochs': [10,20]
}

# Hyperparameter tuning for each model
results = {}
best_models = {}

for model_type in ['RNN', 'LSTM', 'GRU']:
    print(f"\nTuning hyperparameters for {model_type}...")
    best_val_rmse = float('inf')
    best_model = None
    best_params = None

    for learning_rate in hyperparams['learning_rate']:
        for dropout_rate in hyperparams['dropout_rate']:
            for batch_size in hyperparams['batch_size']:
                for epochs in hyperparams['epochs']:
                    print(f"Training {model_type} with lr={learning_rate}, dropout={dropout_rate}, batch_size={batch_size}, epochs={epochs}...")
                    model, _ = build_and_train_model(
                        model_type, (look_back, X_train.shape[2]), X_train, y_train, X_val, y_val, 
                        learning_rate=learning_rate, dropout_rate=dropout_rate, epochs=epochs, batch_size=batch_size
                    )
                    
                    # Evaluating on validation set
                    val_predictions = model.predict(X_val)
                    val_rmse = np.sqrt(mean_squared_error(y_val, val_predictions))
                    
                    # Checking if this is the best model
                    if val_rmse < best_val_rmse:
                        best_val_rmse = val_rmse
                        best_model = model
                        best_params = {
                            'learning_rate': learning_rate,
                            'dropout_rate': dropout_rate,
                            'batch_size': batch_size,
                            'epochs': epochs
                        }

    print(f"Best parameters for {model_type}: {best_params}")
    print(f"Best validation RMSE: {best_val_rmse:.4f}")
    best_models[model_type] = best_model

    # Evaluating the best model on the test set
    train_predictions = best_model.predict(X_train)
    val_predictions = best_model.predict(X_val)
    test_predictions = best_model.predict(X_test)

    # Compute metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))
    val_rmse = np.sqrt(mean_squared_error(y_val, val_predictions))
    test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))
    train_mse = mean_squared_error(y_train, train_predictions)
    val_mse = mean_squared_error(y_val, val_predictions)
    test_mse = mean_squared_error(y_test, test_predictions)
    test_mape = mean_absolute_percentage_error(y_test, test_predictions) * 100
    test_accuracy = 100 - test_mape
    test_r2 = r2_score(y_test, test_predictions)

    results[model_type] = {
        'Train RMSE': train_rmse,
        'Val RMSE': val_rmse,
        'Test RMSE': test_rmse,
        'Train MSE': train_mse,
        'Val MSE': val_mse,
        'Test MSE': test_mse,
        'Test MAPE': test_mape,
        'Test Accuracy (%)': test_accuracy,
        'Test R²': test_r2
    }

    # Plot individual graph for each model
    plt.figure(figsize=(14, 8))
    total_actual = np.concatenate([y_train, y_val, y_test])
    total_predictions = np.concatenate([train_predictions, val_predictions, test_predictions])
    plt.plot(range(len(total_actual)), total_actual, label='Actual Values', color='black', linewidth=2)
    plt.plot(range(len(train_predictions)), train_predictions, label='Training Predictions', color='blue')
    plt.plot(range(len(train_predictions), len(train_predictions) + len(val_predictions)), val_predictions, label='Validation Predictions', color='orange')
    plt.plot(range(len(train_predictions) + len(val_predictions), len(total_actual)), test_predictions, label='Testing Predictions', color='green')
    plt.title(f"{model_type} - Actual vs Predicted Stock Prices (Normalized Scale)")
    plt.xlabel("Time Steps")
    plt.ylabel("Normalized Value")
    plt.legend()
    plt.grid()
    plt.show()

# Print summary results
print("\nModel Performance Summary:")
for model_type, metrics in results.items():
    print(f"\n{model_type} Performance:")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value:.4f}")
```
