```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers, datasets
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import ParameterGrid
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load CIFAR-10 Data
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Data Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

# ResNet Block with L2 Regularization
def resnet_block(inputs, filters, stride=1, regularizer=regularizers.l2(0.001)):
    x = layers.Conv2D(filters, 3, strides=stride, padding="same", use_bias=False, kernel_regularizer=regularizer)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Conv2D(filters, 3, padding="same", use_bias=False, kernel_regularizer=regularizer)(x)
    x = layers.BatchNormalization()(x)
    
    shortcut = inputs
    if stride != 1 or inputs.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, 1, strides=stride, use_bias=False, kernel_regularizer=regularizer)(inputs)
        shortcut = layers.BatchNormalization()(shortcut)
    
    x = layers.Add()([x, shortcut])
    return layers.ReLU()(x)

# ResNet-18 Model with Optimized Dropout and Regularization
def build_resnet18(input_shape=(32, 32, 3), num_classes=10):
    inputs = layers.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = layers.Conv2D(32, 3, strides=1, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = resnet_block(x, 32)
    x = resnet_block(x, 32, stride=2)
    x = resnet_block(x, 64)
    x = resnet_block(x, 64, stride=2)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(0.001))(x)
    
    model = models.Model(inputs, outputs)
    return model

# AlexNet Model with Optimized Dropout and Regularization
def build_alexnet(input_shape=(32, 32, 3), num_classes=10):
    model = models.Sequential([
        data_augmentation,
        layers.Conv2D(32, kernel_size=5, strides=1, padding="same", activation='relu', input_shape=input_shape, kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=2, strides=2),
        layers.Conv2D(64, kernel_size=5, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=2, strides=2),
        layers.Conv2D(128, kernel_size=3, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Flatten(),
        layers.Dropout(0.4),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(0.001))
    ])
    return model

# MobileNet Model with Optimized Dropout and Regularization
def build_mobilenet(input_shape=(32, 32, 3), num_classes=10):
    base_model = MobileNet(weights=None, input_shape=input_shape, classes=num_classes)
    model = models.Sequential([
        data_augmentation,
        base_model,
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(0.001))
    ])
    return model

# Function to compile, train, evaluate, and compute confusion matrix parameters
def compile_train_evaluate(model, learning_rate, model_name):
    model.compile(optimizer=optimizers.Adam(learning_rate=learning_rate), 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])
    
    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1, min_lr=1e-5)
    
    history = model.fit(x_train, y_train, epochs=30, batch_size=64, 
                        validation_split=0.2, callbacks=[early_stopping, lr_scheduler], verbose=2)
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
    # Evaluate on test set
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f'{model_name} Test Accuracy: {test_accuracy:.4f}')
    
    # Confusion matrix and metrics
    y_pred = np.argmax(model.predict(x_test), axis=1)
    y_test_labels = y_test.flatten()
    cm = confusion_matrix(y_test_labels, y_pred)
    
    # Calculate confusion matrix parameters
    tp = np.diag(cm)  # True Positives for each class
    fp = np.sum(cm, axis=0) - tp  # False Positives for each class
    fn = np.sum(cm, axis=1) - tp  # False Negatives for each class
    tn = np.sum(cm) - (fp + fn + tp)  # True Negatives for each class
    
    # Calculate additional metrics
    precision = np.mean(tp / (tp + fp + np.finfo(float).eps))  # Prevent division by zero
    recall = np.mean(tp / (tp + fn + np.finfo(float).eps))
    f1_score = 2 * (precision * recall) / (precision + recall + np.finfo(float).eps)
    
    # Store metrics in dictionary format
    metrics = {
        "Model": model_name,
        "Test Accuracy": test_accuracy,
        "True Positives (TP)": tp.sum(),
        "False Positives (FP)": fp.sum(),
        "False Negatives (FN)": fn.sum(),
        "True Negatives (TN)": tn.sum(),
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score
    }
    return metrics

# Hyperparameter tuning
hyperparameters = {
    'learning_rate': [0.001, 0.0005],
}

# Results for each model and hyperparameter combination
results = []
for params in ParameterGrid(hyperparameters):
    lr = params['learning_rate']
    
    print(f"Training ResNet-18 with learning_rate={lr}")
    resnet18 = build_resnet18()
    resnet18_metrics = compile_train_evaluate(resnet18, lr, f"ResNet-18 (lr={lr})")
    results.append(resnet18_metrics)
    
    print(f"\nTraining AlexNet with learning_rate={lr}")
    alexnet = build_alexnet()
    alexnet_metrics = compile_train_evaluate(alexnet, lr, f"AlexNet (lr={lr})")
    results.append(alexnet_metrics)
    
    print(f"\nTraining MobileNet with learning_rate={lr}")
    mobilenet = build_mobilenet()
    mobilenet_metrics = compile_train_evaluate(mobilenet, lr, f"MobileNet (lr={lr})")
    results.append(mobilenet_metrics)

# Create a DataFrame to display the comparison table
results_df = pd.DataFrame(results)
print("\nConfusion Matrix and Additional Metrics Comparison:")
print(results_df)

# Optional: Save the results to a CSV file for further analysis
results_df.to_csv("model_comparison_results.csv", index=False)

# Display the results in a structured table for clarity
print("\nSummary of Model Performance:")
print(results_df)
```
